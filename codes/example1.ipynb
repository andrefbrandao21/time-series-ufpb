```r
library("forecast")
```

```r
y = arima.sim(list(ma= c(0, 0)), n=1000)
plot(y)
Acf(y)
Pacf(y)
```

```r
y = arima.sim(list(ma= c(0, 8)), n=1000)
plot(y)
Acf(y)
Pacf(y)
```

```r
y = arima.sim(list(ma= c(0,6, 0.4)), n=1000)
plot(y)
Acf(y)
Pacf(y)
```

```r
y = arima.sim(list(ar= c(0.9)), n=1000)
plot(y)
Acf(y)
Pacf(y)
```

```r
y = arima.sim(list(ar= c(0.99)), n=1000)
plot(y)
Acf(y)
Pacf(y)
```

```r
y = arima.sim(list(ar= c(0.45, 0.5)), n=1000)
plot(y)
Acf(y)
Pacf(y)
```

```r
fit = arima(y, order = c(2, 0, 0))
fit
```

```r
library("readxl")
```

```r
y = read_excel('Figure6_5_Table6_1_treasury.xls')

# selecionando apenas a coluna de interesse
y=y[,2]

#transformando a série em um objeto ts
y=ts(y)

plot(y)
Acf(y)
Pacf(y)
```

```r
fit = arima(y, order = c(0,0,1))
fit
```
